# DialFRED Text Baseline

Reproduce the proof-of-concept baseline: download DialFRED, aggregate instruction-level labels, embed instructions with `all-MiniLM-L6-v2`, and train the MLP ambiguity classifier.

```python
from sensor_vlm.data import load_instruction_labels, print_dataset_summary

labels = load_instruction_labels()
print_dataset_summary(labels)
labels.head()
```

```python
from sensor_vlm.features import build_text_baseline_cache
from sensor_vlm.paths import FEATURES_DIR

cache_path = FEATURES_DIR / "dialfred_text_baseline.npz"
build_text_baseline_cache(labels, cache_path)
cache_path
```

```python
from sensor_vlm.train import train_from_cache
from sensor_vlm.paths import CHECKPOINTS_DIR, REPORTS_DIR

train_from_cache(
    cache_path,
    checkpoint_path=CHECKPOINTS_DIR / "best_text_mlp.pt",
    report_path=REPORTS_DIR / "text_baseline_report.txt",
    epochs=50,
)
```
