# BLIP-2 Image Features

Run frozen BLIP-2 on a local image or URL and inspect captions, VQA answers, caption variance, and Q-Former embeddings.

```python
from sensor_vlm.blip2_features import Blip2FeatureExtractor

image_path = "https://storage.googleapis.com/sfr-vision-language-research/BLIP/demo.jpg"
extractor = Blip2FeatureExtractor()
features = extractor.extract_features(image_path)

print("Caption:", features.caption)
print("Caption ambiguity score:", features.caption_ambiguity_score)
print("Q-Former pooled shape:", features.qformer_pooled.shape)
```
